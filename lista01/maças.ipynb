```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
q_g = 38.9  # W * m-3
h = 7.5  # W * m-2 * k-1
T_inf = 278  # K
Kappa = 0.5  # W * m-1 * -K
R = 0.04 # m
```

```python
# Domnio
r = np.linspace(0, R, 1000)

# Temperatura radial
T_r = (-q_g * r**2) / (6 * Kappa) + T_inf + (q_g * R) / 3 * ((1/h) - (R)/(2 * Kappa))

# Temperaturas nos pontos de interesse
T_center = T_r[0]
T_surface = T_r[-1]

# Estilo do grfico
sns.set_theme(style="whitegrid")

plt.figure(figsize=(15, 8))
sns.lineplot(x=r, y=T_r, color='crimson', linewidth=2.5, label="Distribuio de Temperatura")

# Pontos azuis em r=0 e r=R
plt.scatter([0, R], [T_center, T_surface], color='blue', s=80, zorder=5, label="Pontos r = 0 e r = R")

# Anotaes prximas aos pontos
plt.annotate(f"T(0) = {T_center:.2f} K", xy=(0, T_center), xytext=(0, T_center - 0.002),
             arrowprops=dict(arrowstyle="->", color='blue'), fontsize=12, color='blue', ha='left')

plt.annotate(f"T(R) = {T_surface:.2f} K", xy=(R, T_surface), xytext=(R - 0.002, T_surface + 0.001),
             arrowprops=dict(arrowstyle="->", color='blue'), fontsize=12, color='blue', ha='right')

# Eixos e ttulo
plt.title("Distribuio de Temperatura da Ma", fontsize=20, fontweight='bold')
plt.xlabel("Raio (m)", fontsize=16)
plt.ylabel("Temperatura (K)", fontsize=16)
plt.ticklabel_format(useOffset=False, style='plain', axis='y')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
